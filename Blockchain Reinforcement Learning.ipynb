```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee','grid'])
# 提供的数据
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
research_papers = [1, 5, 16, 43, 106, 111, 100, 2]

# 创建折线图
plt.plot(years, research_papers, marker='o', linestyle='-', color='b')

# 添加标题和标签
plt.title('Blockchain and Reinforcement Learning Research Papers Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Papers')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use(['science'])
# 提供的新数据
categories = ['Article', 'Meeting', 'Patent', 'Early Access', 'Review Article', 'Dissertation Thesis', 'Other']
counts = [300, 85, 48, 19, 14, 9, 1]

# 创建饼图
plt.figure(figsize=(5, 5))
plt.pie(counts, labels=categories, autopct='%1.1f%%', startangle=140)

# 添加标题
plt.title('Distribution of Research Output Types')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 提供的新数据
categories = ['Article', 'Meeting', 'Patent', 'Early Access', 'Review Article', 'Dissertation Thesis', 'Other']
counts = [300, 85, 48, 19, 14, 9, 1]
# 设置每种类型文章的颜色
colors = ['blue', 'green', 'orange', 'red', 'purple', 'brown', 'gray']

# 创建柱状图
plt.bar(categories, counts, color=colors)
# 创建柱状图
# plt.bar(categories, counts, color='blue')

# 添加标题和标签
plt.title('Distribution of Research Output Types')
plt.xlabel('Research Output Types')
plt.ylabel('Number of Papers')

# 调整横轴标签的显示
plt.xticks(rotation=45, ha='right', fontsize=5)

# 显示图表
plt.show()
```
